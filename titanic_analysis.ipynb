```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic_df = pd.read_csv('./titanic_train.csv')
# Dataframe head
titanic_df.head(3)
```

```python
# Dataframe informations
titanic_df.info()
```

```python
# Dataframe distributions
titanic_df.describe()
```

```python
# Dataframe column value count
# Use value_counts() method in Series type
# Use dropna=False to count Null values
titanic_df['Pclass'].value_counts(dropna=False)
```

```python
# Dataframe Drop
# label=index, axis=0: row drop
# label=column name, axis=1: column drop
titanic_df.drop(labels='Fare', axis=1)
```

```python
# Index type
# Same as PK in database
# Cannot be changed
titanic_df.index
```

```python
# Use reset_index to reset original index
# drop=True deletes original index
titanic_df.reset_index(drop=True)
```

```python
# Indexing only can use columns
titanic_df[['Name', 'Age']].head(3)
```

```python
# iloc: get data by row, column number
titanic_df.iloc[1, 4]
```

```python
# loc: get data by row, column name
titanic_df.loc[1, 'Age']
```

```python
# Can use boolean indexing on loc method
titanic_df.loc[titanic_df['Age']>60, ['Name', 'Age']].head(3)
```

```python
# Can use multiple boolean conditions
# and: &, or: |, not: ~
titanic_df.loc[(titanic_df['Age']<60) & (titanic_df['Age']>20), ['Name', 'Age']].head(3)
```

```python
# Use sort_values() to sort DataFrame or Series
titanic_sorted = titanic_df.sort_values(by='Name')
titanic_sorted.head(3)
```

```python
# Aggregation methods: min, max, sum, count ...
# Aggregation methods don't calculate Null values
titanic_df[['Age', 'Pclass']].mean()
```

```python
# groupby(by='column') to group DataFrame by columns
# Then aggregate DataFrame
# Can do column indexing
titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
```

```python
# Need to use agg() to get multiple aggregations
# Use aggregation keywords
titanic_df.groupby('Pclass')['Age'].agg([min, max])
```

```python
# Check Null values
titanic_df.isna().head(3)
```

```python
# Change Null values to other values
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
# Processing data by apply(lambda) methods
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
# Processing data by apply(lambda) methods
# method can be declared first and use later
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(5)
```
